```python
import numpy as np
import matplotlib.pyplot as plt

def chebyshev2(A, max_k=5):
    eigvals = np.linalg.eigvals(A)
    l_n = max(abs(eigvals))
    l_1 = min(abs(eigvals))

    d = (l_n + l_1) / 2
    c = (l_n - l_1) / 2

    lambdas = np.linspace(l_1, l_n, 1000)
    max_vals = []

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

    
    for k in range(1, max_k + 1):
        Pk = np.ones_like(lambdas)
        for i in range(1, k + 1):
            omega_i = 1 / (d - c * np.cos((2 * i - 1) * np.pi / (2 * k)))
            Pk *= (1 - omega_i * lambdas)
        ax1.plot(lambdas, np.abs(Pk), label=f"k={k}")

        max_val_k = np.max(np.abs(Pk))
        max_vals.append(max_val_k)

    ax1.set_title("Polynomy pro k = 1 až 5")
    ax1.set_xlabel("λ")
    
    ax1.grid(True)
    ax1.legend()

    
    ax2.plot(range(1, max_k + 1), max_vals, marker='o')
    ax2.set_title("Konvergence maxima")
    ax2.set_xlabel("k")
    
    ax2.grid(True)

    plt.tight_layout()
    plt.show()


A = np.array([
    [ 7.015,  0.551,  0.794, -0.368,  0.889],
    [ 0.551,  4.319,  0.467, -0.066,  0.458],
    [ 0.794,  0.467,  5.738, -0.036,  0.724],
    [-0.368, -0.066, -0.036,  2.495, -0.187],
    [ 0.889,  0.458,  0.724, -0.187,  3.432]
])

chebyshev2(A)
```
